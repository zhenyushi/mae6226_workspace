```python
import numpy
import math
from matplotlib import pyplot
%matplotlib inline
```

```python
N = 50
x_start, x_end = -2.0, 2.0          
y_start, y_end = -1.0, 1.0          
x = numpy.linspace(x_start*2, x_end*2, N)    
y = numpy.linspace(y_start*2, y_end*2, N)   
X, Y = numpy.meshgrid(x, y)             
```

```python
gamma = 50
points = 200
R = 50
x_vortex = numpy.linspace(-50., 50, num=points)
y_vortex = numpy.zeros(points)
#y_vortex = numpu.sqrt(r**2 - x**2)
```

```python
def get_velocity_vortex(strength, xv, yv, X, Y):
    
    u = + strength/(2*math.pi)*(Y-yv)/((X-xv)**2+(Y-yv)**2)
    v = - strength/(2*math.pi)*(X-xv)/((X-xv)**2+(Y-yv)**2)
    
    return u, v
```

```python
u_vortex_sum,v_vortex_sum,psi_vortex_sum = 0,0,0
for i in range(len(x_vortex)):
    u_vortex, v_vortex = get_velocity_vortex(gamma, x_vortex[i], y_vortex[i], X, Y)
    u_vortex_sum = u_vortex+ u_vortex_sum
    v_vortex_sum = v_vortex+ v_vortex_sum
    #psi_vortex = get_stream_function_vortex(gamma, x_vortex[i], y_vortex[i], X, Y)
    #psi_vortex_sum = psi_vortex + psi_vortex_sum
```

```python
size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u_vortex_sum, v_vortex_sum, density=1, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.scatter(x_vortex, y_vortex, color='#CD2305', s=20, marker='o');
```
